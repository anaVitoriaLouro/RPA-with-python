### Challenge:

Everyday the company's system updates the previous day's sales. As an analyst, you daily job is to send an e-mail to your superiors with the following informations: the income and how many products were sold total 

- We're gonna solve this problem using 'pyautogui' library
- To test the code you can utilize your own e-mail

```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

# Step 1: Access the company system
pyautogui.hotkey('ctrl', 't')
pyautogui.write('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

time.sleep(5)

# Step 2: Finding the database
pyautogui.click(x=556, y=329, clicks=2)
time.sleep(2)

# Step 3: Downloading the datebase
pyautogui.click(x=473, y=383)
pyautogui.click(x=1668, y=198)
pyautogui.click(x=1411, y=737)
time.sleep(7)
```

### Getting the needed information

- Income
- Total Products Sold

```python
# Step 4: Visualizing the database
import pandas as pd

tabela = pd.read_excel('Vendas - Dez.xlsx')
tabela
```

```python
# Step 5: Calculating the indicators

quantidade = tabela['Quantidade'].sum()
faturamento = tabela['Valor Final'].sum()

print(quantidade)
print(faturamento)
```

### Sending the e-mail

```python
# Step 6: Accessing the e-mail
pyautogui.hotkey('ctrl', 't')
pyautogui.write('https://mail.google.com/mail/u/0/?tab=rm&ogbl')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

time.sleep(5)

# Step 7: Sending the e-mail
pyautogui.click(x=91, y=220)
pyautogui.write('youre_mail@youre_mail.com') # write the e-mail here
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')
pyautogui.press('tab')

# E-mail's subject
pyperclip.copy('Relatório de Vendas')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')


# E-mail's body
texto = f'''
Dear Mr. , good morning

Yesterday's turnover was R${faturamento:,.2f}
And the total sold was {quantidade:,}

Best regards,
Your name
'''

pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')

# Sending the e-mail
pyautogui.hotkey('ctrl', 'enter')
```
